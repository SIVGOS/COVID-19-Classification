Step 3: We created a pre-processing block for the images. It is observed that when the images are pre-processed, the performance slightly enhences. The pre-processing stage highlights the edges of the image through a process similar to unsharp masking. The steps include:
* Find the edges of the images using a custom spatial high pass filter
* Perform gamma correction of the filtered image to improve the contrast
* Superimpose the filtered image (2D) upon the original image (3D)

Finally, the image is resized according to the specified size.

```
import numpy as np
import cv2
import os
def adjust_gamma(image, gamma=1.0):
	# build a lookup table mapping the pixel values [0, 255] to
	# their adjusted gamma values
	invGamma = 1.0 / gamma
	table = np.array([((i / 255.0) ** invGamma) * 255
		for i in np.arange(0, 256)]).astype("uint8")
	# apply gamma correction using the lookup table
	return cv2.LUT(image, table)
 
def rotate_image(image, angle):
  image_center = tuple(np.array(image.shape[1::-1]) / 2)
  rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
  result = cv2.warpAffine(image, rot_mat, image.shape[1::-1], flags=cv2.INTER_LINEAR)
  return result

def distort_image(im):
    # give a random zoom to every image (maximum 20 pixel)
    rz = np.random.randint(0,21)
    [m,n,h] = im.shape
    im2 = cv2.resize(im, (m+rz, n+rz))
    im3 = im2[round(rz/2):round(rz/2)+m,round(rz/2):round(rz/2)+m,:]
    # randomly flip some images
    if np.random.rand(1)>0.5:
        im3 = cv2.flip(im3,1)
    # give a random rotation to some of the images
    if np.random.rand(1)>0.5:
      ang = np.random.randint(-90,91)
      im3 = rotate_image(im3,ang)
    return im3

def load_images_from_folder(folder, xsize, ysize, batch_size=0, distort = False):
    all_files = os.listdir(folder)
    if batch_size > 0 and batch_size < len(all_files):
      np.random.shuffle(all_files)
      all_files = all_files[:batch_size]
    images = np.zeros([len(all_files),xsize,ysize,3])
    cnt = 0
    for filename in all_files:
        img1 = cv2.imread(os.path.join(folder,filename))
        if img1 is not None:
          img2 = adjust_gamma(img1,gamma=1.7)
          img_resized=cv2.resize(img2, (xsize,ysize))
          if distort:
            img_dist = distort_image(img_resized)
            images[cnt] = img_dist.astype('float32')/255.0
          else:
            images[cnt] = img_resized.astype('float32')/255.0
          cnt+=1
    return images
```

Since our RAM is limited, we decided not to load the entire dataset at once. The size of the features are very small and resizing the images degrades the quality significantly. Hence, a custom data generator is used. The following section creates the custom data generator.

We know that a datagenerator should return the data iteratively and hence the we use the keyword "yield" instead of "return". Another data generator function (datagen_static) is written which is not iterative. The purpose of this function is to generate a static set of data for validation and/or testing as required.

```
from sklearn.utils import shuffle
from tensorflow.keras.utils import to_categorical
xsize, ysize = 299,299
# iterative data generator
def generator(base_path, batch_size=64, distort = False):
  global xsize, ysize
  n0 = round(batch_size*0.58)
  n1 = round(batch_size*0.39)
  n2 = round(batch_size*0.03)
  while n0+n1+n2 > batch_size:
    n0-=1
  while True:
    img_0 = load_images_from_folder(os.path.join(base_path,'0'),xsize,ysize,batch_size=n0, distort=distort)
    img_1 = load_images_from_folder(os.path.join(base_path,'1'),xsize,ysize,batch_size=n1, distort=distort)
    img_2 = load_images_from_folder(os.path.join(base_path,'2'),xsize,ysize,batch_size=n2, distort=distort)
    images = np.concatenate((img_0, img_1, img_2), axis=0)
    labels = to_categorical(np.concatenate((np.zeros(n0), np.ones(n1), 2*np.ones(n2))), 3)
    X_train, Y_train = shuffle(images, labels)
    yield X_train, Y_train
```

```
w0, w1, w2 = 1/(3*0.58), 1/(3*0.39), 1/(3*0.03)
```

Now let's load tensorflow and keras packages to create the Sequential CNN model

```
import tensorflow
from tensorflow.keras.models import Sequential, Model
from tensorflow.keras.layers import Dense, Dropout, Conv2D, AveragePooling2D, MaxPooling2D, Flatten, Activation, GlobalAveragePooling2D, BatchNormalization, Input
from tensorflow.keras.optimizers import RMSprop
from tensorflow.keras.applications import VGG16,InceptionV3,ResNet50, MobileNetV2
from tensorflow.keras.optimizers import Adam
```

Load the Inception V2 imagenet as base model

```
basemodel = InceptionV3(weights="imagenet", include_top=False,input_tensor=Input(shape=(xsize, ysize, 3)))
```

```
n = len(basemodel.layers)
nt = 20 # number of layers to train
for k in range(n-nt):
  basemodel.layers[k].trainable = False
```

```
headModel = basemodel.output
headModel = AveragePooling2D(pool_size=(2, 2))(headModel)
headModel = Flatten(name="flatten")(headModel)
headModel = Dropout(0.5)(headModel)
headModel = Dense(181, activation="relu")(headModel)
headModel = Dropout(0.3)(headModel)
headModel = Dense(3, activation="softmax")(headModel)
```

Compile the model

```
model = Model(inputs=basemodel.input, outputs=headModel)
```

```
 model.summary()
```

Define a model check point to programmatically save the model whenever the validation accuracy is the minimum.

```
epochs= 200
lr = 1e-4
```

```
model.compile(loss="categorical_crossentropy", optimizer=Adam(lr=1e-4,decay=lr/epochs),metrics=["accuracy"])
```

```
from tensorflow.keras.callbacks import ModelCheckpoint
mc = ModelCheckpoint('best_model.h5', monitor='val_loss', mode='min', verbose=1, save_best_only=True)
```

```
train_gen = generator('training/train_data',batch_size=128, distort = True)
val_gen = generator('training/val_data', batch_size=128, distort = False)
```

Start training the model (Don't run this without a GPU)

```
history = model.fit(train_gen, #train_images, train_labels,
                    steps_per_epoch = 10,
                    epochs=epochs,
                    verbose=2,
                    validation_data=val_gen,
                    validation_steps=5,
                    class_weight = {0: w0, 1: w1, 2: w2},
                    callbacks=[mc])
```

```
from tensorflow.keras.models import load_model
test_images, test_labels = next(generator('training/test_data',batch_size=1000))
print('With best model')
best_model = load_model('best_model.h5')
score = best_model.evaluate(test_images, test_labels, verbose=0)
print('Test loss:', score[0])
print('Test accuracy:', score[1])

print('With final model')
score1 = model.evaluate(test_images, test_labels, verbose=0)
print('Test loss:', score1[0])
print('Test accuracy:', score1[1])
```
